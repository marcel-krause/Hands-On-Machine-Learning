# Image Clustering

```python
from matplotlib.image import imread
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## Load the image

First, we load an image.<br><br>
*Image source*: https://commons.wikimedia.org/w/index.php?curid=21175158<br>
*Author*: Diego Delso<br>
*License*: CC BY-SA 3.0

```python
image = imread('field.jpg')
image.shape
```

```python
X = image.reshape(-1, 3)
```

```python
plt.figure(figsize=(image.shape[0]/100,image.shape[1]/100))
plt.imshow(image)
```

## Train the K-means model

### 10 colors

```python
kmeans = KMeans(n_clusters=10).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]/255.
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.figure(figsize=(image.shape[0]/100,image.shape[1]/100))
plt.imshow(segmented_img)
```

### 8 colors

```python
kmeans = KMeans(n_clusters=8).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]/255.
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.figure(figsize=(image.shape[0]/100,image.shape[1]/100))
plt.imshow(segmented_img)
```

### 6 colors

```python
kmeans = KMeans(n_clusters=6).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]/255.
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.figure(figsize=(image.shape[0]/100,image.shape[1]/100))
plt.imshow(segmented_img)
```
